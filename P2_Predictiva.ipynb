cantidad de missing values

```python
def cant_missings (df):
  cant_missing = df.isnull().sum()
  missings = cant_missing.sum()
  return missings

print(cant_missings(df))
```

Procedo a eliminar aquellas columnas con missing values mayor al 90%

```python
def per60 (df):
  porcentajes = (df.isnull().sum()/len(df))*100
  cols = porcentajes.keys()
  dropcol = []
  for col in cols:
    if porcentajes[col] > 60:
      dropcol.append(col)
  return dropcol

df = df.drop(columns = per60(df), axis=1)
```

**Directors y Writers**

Son atributos relevantes los cuales tendrán el mismo tratamiento siéndo de gran utilidad no solo la variable target, pero también minutos vistos y votos al ser altamente relacionadas con el target. 

```python
def Lista_Data (df,target,sep):
    def Lista_Fila (row,Facum,target,sep):
        Fsp = row[target].split(sep)
        for i in Fsp:
            nuevasF = row.to_dict()
            nuevasF[target] = i
            Facum.append(nuevasF)
    nuevasF = []
    df.apply(Lista_Fila,axis=1,args = (nuevasF,target,sep))
    df_nuevo = pd.DataFrame(nuevasF)
    return df_nuevo
```

```python
df["directors"] = df["directors"].astype("str")
tablaDir = Lista_Data(df[["directors","averageRating","runtimeMinutes", "numVotes"]],"directors",",")
df["writers"] = df["directors"].astype("str")
tablaWri = Lista_Data(df[["writers","averageRating","runtimeMinutes", "numVotes"]],"writers",",")
```
